```python
## Comenzamos importando las librerias a utilizar

import os
import numpy as np
import pandas as pd

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None) ## Muestra todas las columnas del dataset
```

```python
## Cargamos el archivo que contiene los valores de nuestro dataset

df = pd.read_csv(os.path.join("in/fifa.csv"))
df.head(10)
```

```python
## Miramos que columnas tienen un valor desconocido

print(df.columns[df.isna().any()].tolist())
```

```python
df[df.isna()["Loaned From"]==False].head(10)
```

```python
## Nos piden saber de qué club es un jugador por su valor económico. Por lo tanto, todas aquellas columnas
## con imágenes como valor no nos servirán, así que las eliminamos. Además, el ID del jugador dentro del 
## juego no será muy relevante para nuestro objetivo, así que también prescindiremos de este campo. También
## eliminaremos aquellos jugadores sin club actualmente según el dataset

df = df.dropna(subset=['Club'])

df.drop(["Unnamed: 0","ID","Photo", "Flag", "Club Logo", "Special", "Loaned From"], axis = 1, inplace = True)
df
```

```python
## Una vez hemos eliminado aquellas columnas que no nos sirven para predecir el valor, normalizaremos varias
## características, agrupándolas por rangos numéricios. En este caso, solamente agruparemos las edades por 
## rangos, diciendo si un jugador es joven, veterano o de rango medio

n_datos = [df]

for dataset in n_datos:
    dataset['Age range'] = pd.cut(pd.to_numeric(dataset['Age']), bins=[0,20,30,50], labels=['Rookie', 'Normal', 'Veteran'])

## Los porteros incluyen mucho valores nulos en cada columna que les pertoca, así que también 
## serán eliminados de muestro dataset
df = df[df.Position != "GK"]

df.head(10)
```

```python
## Para convertir todos los valores económicos a un número legible por la máquina, definiremos la función
## value_to_float

def value_to_float(x):
    """
    From K and M to float.
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

```python
## Nos interesa obtener como resultado el valor de la columna "Value", respecto a ciertas características.
## Algunas de éstas son diferentes, dependiendo por ejemplo de si el jugador es portero o no.
## Para solventar este problema, aunque haya muchas más columnas de las deseadas, usaremos la función pd.get_dummies

df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)

df = pd.get_dummies(df, columns=['Nationality','Club','Position', 'Age range'],prefix=['Nationality','Club','Position', 'Age range'])

df.head(10)
```

```python
## Ahora, empezamos con el proceso de predicción. Para ello, fijareos una columna

name = df.pop("Name")
val = df.pop("Value")
df = df.drop([c for c in df.columns if not c.startswith(('Age range', 'Nationality', 'Club', 'Overall', 'Potential', 'Wage'))], axis=1)

df.head(10)
```

```python
## Preparamos nuestros conjuntos de entrenamiento y pruebas

X_train, X_test, Y_train, Y_test = train_test_split(df, val, test_size=0.25, random_state=42)

len(X_train)
```

```python
## Entremanos nuestro conjunto de entrenamiento con un modelo de regresión lineal

from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(X_train, Y_train)
```

```python
## Predecimos los valores con nuestro conjunto de pruebas

predsTest = reg.predict(X_test)
print("predsTest[0] = " + str(predsTest[0]))
```
